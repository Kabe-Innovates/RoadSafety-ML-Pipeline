## 1. Executive Summary

### Project Objective
Build a machine learning pipeline to predict traffic accident occurrence using 840 traffic data samples with 14 features.

### Methodology
- **Exploratory Data Analysis**: Examined data distributions, missing values (42 per column), and feature relationships
- **Data Preprocessing**: Imputed missing values (median for numeric, mode for categorical)
- **Feature Engineering**: Scaled numeric features and one-hot encoded categorical features
- **Modeling**: Trained 4 foundational models (Logistic Regression, Decision Tree, Random Forest, SVM)
- **Evaluation**: Compared models using accuracy, precision, recall, F1, and ROC-AUC metrics

````python
# Load and display metrics
metrics_df = pd.read_csv(RESULTS_DIR / "model_metrics.csv", index_col='Model')
metrics_df
````

## 2. Key Findings

````python
# Best model analysis
best_by_f1 = metrics_df['F1'].idxmax()
best_by_accuracy = metrics_df['Accuracy'].idxmax()

print(f"Best Model (by F1): {best_by_f1}")
print(f"  F1 Score: {metrics_df.loc[best_by_f1, 'F1']:.4f}")
print(f"  Accuracy: {metrics_df.loc[best_by_f1, 'Accuracy']:.4f}")
print(f"  ROC-AUC: {metrics_df.loc[best_by_f1, 'ROC-AUC']:.4f}")

print(f"\nBest Model (by Accuracy): {best_by_accuracy}")
print(f"  Accuracy: {metrics_df.loc[best_by_accuracy, 'Accuracy']:.4f}")
````

## 3. Data Insights

### Missing Values
- Every column had exactly 42 missing values
- Total dataset: 840 rows × 14 columns
- 5% missing data per column
- Handling: Imputation using median (numeric) and mode (categorical)

### Features
- **6 Numeric Features**: Traffic_Density, Speed_Limit, Number_of_Vehicles, Driver_Alcohol, Driver_Age, Driver_Experience
- **7 Categorical Features**: Weather, Road_Type, Time_of_Day, Accident_Severity, Road_Condition, Vehicle_Type, Road_Light_Condition
- **Target**: Accident (binary classification)

### Feature Engineering
- Numeric features: StandardScaler applied
- Categorical features: One-hot encoded
- Final feature matrix: 6 numeric + encoded categorical = [total] features
