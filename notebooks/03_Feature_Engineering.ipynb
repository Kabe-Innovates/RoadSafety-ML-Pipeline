```python
# Import Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
# Load Dataset
df = pd.read_csv("../datasets/preprocessed_traffic_accident_prediction.csv")
target = pd.read_csv("../datasets/target_traffic_accident_prediction.csv")

df.head()
```

```python
# Split Numerical and Categorical Features based on types
num_cols = df.select_dtypes(include=[np.number]).columns
nom_cols = df.select_dtypes(include=["str"]).columns
```

```python
# Encode nominal features using OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, drop="first")
encoded_nom = encoder.fit_transform(df[nom_cols])
encoded_nom_df = pd.DataFrame(encoded_nom, columns=encoder.get_feature_names_out(nom_cols))
```

```python
# Scale numerical features using StandardScaler
scaler = StandardScaler()
scaled_num = scaler.fit_transform(df[num_cols])
scaled_num_df = pd.DataFrame(scaled_num, columns=num_cols)
```

```python
# Combine scaled numerical and encoded nominal features
X = pd.concat([scaled_num_df, encoded_nom_df], axis=1)
X.head()

#Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, target, test_size=0.2, random_state=42)

# Save split datasets
X_train.to_csv("../datasets/X_train.csv", index=False)
X_test.to_csv("../datasets/X_test.csv", index=False)
y_train.to_csv("../datasets/y_train.csv", index=False)
y_test.to_csv("../datasets/y_test.csv", index=False)
```
